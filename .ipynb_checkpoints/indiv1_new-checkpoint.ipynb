```python
def plt_show(image, color = None):
    pic_size=5
    plt.figure(figsize = (pic_size, pic_size))
    plt.imshow(image, color)
    plt.show()
```

```python
pic_size = 5
ref = cv.imread("pic1.jpg")
plt.figure(figsize = (pic_size, pic_size))
plt.imshow(ref)
```

```python
bin_ref = cv.blur(cv.cvtColor(ref, cv.COLOR_BGR2GRAY), (10,10))
plt_show(bin_ref, 'gray')
```

```python
# for c in range(50,150,10):
#         for d in range(60,180,10):
#             print(c,d)
#             bin_ref_canny = cv.Canny(bin_ref, c, d)
#             plt_show(bin_ref_canny, "gray")
                
```

```python
bin_ref_canny = cv.Canny(bin_ref, 50, 170)
plt_show(bin_ref_canny, "gray")
```

```python
circle = cv.getStructuringElement(cv.MORPH_ELLIPSE, (15,10))
img = cv.dilate(bin_ref_canny, circle)
```

```python
cv.floodFill(img, None,(0,0), newVal=(255, 255, 255))
plt_show(img, "gray")
```

```python
#kernel = np.ones((3,3),np.uint8)
#img = cv.morphologyEx(~img, cv.MORPH_CLOSE,kernel, iterations = 1)
#plt_show(img, 'gray')
```

```python
image_shape = img.shape
for i in range(image_shape[0]):
    for j in range(image_shape[1]):
        if(i==0)|(j==0)|(i==image_shape[0]-1)|(j==image_shape[1]-1):
            img[i][j]=255
plt_show(img, "gray")
```

```python
contours, _ = cv.findContours(img,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
contour=[]
for i in contours:
    if(len(i)>40):
        contour.append(i)
print("Count  = ", len(contour))

# for i in range(46):
#     img0=np.zeros(img.shape)
#     figures_border = cv.drawContours(img0,contour[i],-1,(14,28,57), thickness=10)
#     plt_show(figures_border)
img0=np.zeros(img.shape)
figures_border = cv.drawContours(img0,contour,-1,(14,28,57), thickness=10)
plt_show(figures_border)

figures_border = cv.drawContours(ref,contour,-1,(255,0,0), thickness=10)
plt_show(figures_border)

```
